````python
# -*- coding: utf-8 -*-
"""
Evaluación de respuestas LLM para interpretación de features SAE (solo-positivos).

Qué hace:
- Lee llm-responses.jsonl y detecta los grupos presentes por (sae_model, topk).
- Parsea robustamente objetos {"explanation","activating_sentences"} del LLM.
- Limpia <<...>>(score) y decodifica placeholders [NL],[NL2],[TAB],[WSP] cuando aplica.
- Tokeniza y obtiene residual hasta TARGET_LAYER con el backbone (MLX).
- Calcula activaciones (máximo por oración) vía SAE y usa τ_k desde scan_state_{sae}.pkl.
- Reporta métricas por feature y resúmenes por (SAE,k).
- Realiza comparaciones con prueba t (Welch) de TODOS los SAEs contra el primero (baseline),
  por cada k y también agregando todos los k (ALL k).

Requisitos:
- numpy, scipy, tqdm, mlx, mlx_lm
- Tu clase SparseAutoencoder (clip_norms, encoder) y backbone compatible con mlx_lm.load
"""

from __future__ import annotations
import re
import csv
import json
from pathlib import Path
from typing import Any, Dict, List, Tuple, Iterable, DefaultDict, NamedTuple
from collections import defaultdict

import numpy as np
from tqdm.auto import tqdm
from scipy import stats

import mlx.core as mx
from mlx_lm import load as mlx_load
from model import SparseAutoencoder
from mlp import SparseAutoencoder as SparseAutoencoder_mlp


# =============================================================================
# Configuración
# =============================================================================

ROOT = Path("/Users/josue/proyectos/tesis/lrq-sae").resolve()
INTERP_DIR = ROOT / "interpretability"
DATA_DIR = INTERP_DIR / "data"
RESULTS_JSONL = DATA_DIR / "llm-responses.jsonl"

# SAEs a comparar (mismo orden que en llm-responses y scan_state_*.pkl)
SAE_PATHS = [
    'checkpoints/bsae_512_2048_512__Er0_Dr0_k16_ep16_20251129-040858.pkl',
    'checkpoints/bsae_512_2048_512__Er1_Dr0_k16_ep16_20251129-152503.pkl',
    'checkpoints/bsae_512_2048_512__Er2_Dr0_k16_ep16_20251129-183808.pkl',
    'checkpoints/bsae_512_2048_512__Er0_Dr1_k16_ep16_20251129-204654.pkl',
    'checkpoints/bsae_512_2048_512__Er1_Dr1_k16_ep16_20251129-233024.pkl',
    'checkpoints/bsae_512_2048_512__Er2_Dr1_k16_ep16_20251130-025537.pkl',
    'checkpoints/bsae_512_2048_512__Er0_Dr2_k16_ep16_20251130-051124.pkl',
    'checkpoints/bsae_512_2048_512__Er1_Dr2_k16_ep16_20251130-080037.pkl',
    'checkpoints/bsae_512_2048_512__Er2_Dr2_k16_ep16_20251130-113058.pkl',
    'checkpoints/sae_512_2048_512__MLP1024_k16_ep16_20251209-215356.pkl'
]
SAE_LABELS = [Path(p).stem for p in SAE_PATHS]

LLM_PATH = "EleutherAI/pythia-70m-deduped"
TARGET_LAYER = 3
BATCH_SIZE = 64

# Tamaño de población finita: nº total de features por SAE
POP_SIZE_FEATURES = 2048


# -----------------------------------------------------------------------------
# Cargar backbone y tokenizer
# -----------------------------------------------------------------------------
model, tokenizer = mlx_load(LLM_PATH)
if getattr(tokenizer, "pad_token_id", None) is None:
    try:
        tokenizer.pad_token = tokenizer.eos_token
    except Exception:
        pass
if getattr(tokenizer, "pad_token_id", None) is None:
    raise RuntimeError("No se pudo establecer un pad_token_id en el tokenizer.")

# Cargar SAEs
SAES = [SparseAutoencoder.load(str(p)) if 'Er' in p else SparseAutoencoder_mlp.load(str(p)) for p in SAE_PATHS]


# =============================================================================
# Utilidades de texto y JSON (parser robusto, SOLO POSITIVOS)
# =============================================================================

def strip_code_fences(s: str) -> str:
    s = s.strip()
    if s.startswith("```") and s.endswith("```"):
        body = s[3:-3].strip()
        body = re.sub(r"^(json|javascript|js|py|python)\s*\n", "", body, flags=re.IGNORECASE)
        return body.strip()
    return s

def normalize_quotes(s: str) -> str:
    pairs = {
        "\u201c": '"', "\u201d": '"', "\u201e": '"', "\u00ab": '"', "\u00bb": '"',
        "\u2018": "'", "\u2019": "'", "\u201a": "'", "\u2032": "'", "\u2033": '"'
    }
    for k, v in pairs.items():
        s = s.replace(k, v)
    return s

def remove_trailing_commas(s: str) -> str:
    return re.sub(r",(\s*[\}\]])", r"\1", s)

_STR_RE = r'"((?:[^"\\]|\\.)*)"'

def extract_string_field(src: str, key: str) -> str | None:
    m = re.search(rf'"{re.escape(key)}"\s*:\s*{_STR_RE}', src, flags=re.DOTALL)
    return m.group(1) if m else None

def extract_array_of_strings(src: str, key: str) -> List[str] | None:
    m = re.search(rf'"{re.escape(key)}"\s*:\s*\[', src)
    if not m:
        return None
    start = m.end() - 1
    depth, i, end = 0, start, None
    while i < len(src):
        if src[i] == '[':
            depth += 1
        elif src[i] == ']':
            depth -= 1
            if depth == 0:
                end = i + 1
                break
        i += 1
    if end is None:
        return None
    arr_text = src[start:end]
    strings = re.findall(_STR_RE, arr_text, flags=re.DOTALL)
    return [s for s in strings]

def coerce_json_dict_pos_only(d: Dict[str, Any]) -> Dict[str, Any]:
    """
    Normaliza la respuesta del LLM al esquema:
      {"explanation": str, "activating_sentences": List[str]}
    Acepta 'positives' como alias de fallback si aparece accidentalmente.
    """
    out = {"explanation": "", "activating_sentences": []}
    if "explanation" in d:
        out["explanation"] = str(d["explanation"]).strip()
    # clave principal
    if "activating_sentences" in d and isinstance(d["activating_sentences"], list):
        out["activating_sentences"] = [str(x) for x in d["activating_sentences"]]
    # alias por compatibilidad
    elif "positives" in d and isinstance(d["positives"], list):
        out["activating_sentences"] = [str(x) for x in d["positives"]]
    return out

def clamp_k(xs: List[str], k: int = 5) -> List[str]:
    xs = [x.strip() for x in xs if isinstance(x, str) and x.strip()]
    return xs[:k]

# ---- Limpieza de marcadores y placeholders ----

def drop_angle_markers(s: str) -> str:
    return re.sub(r"<<\s*(.*?)\s*>>", r"\1", s, flags=re.DOTALL)

def decode_whitespace_placeholders(s: str) -> str:
    return (s.replace("[NL2]", "\n\n")
             .replace("[NL]", "\n")
             .replace("[TAB]", "\t")
             .replace("[WSP]", " "))

def limpiar_salida_llm(s: str, decode_placeholders: bool) -> str:
    # Quita <<...>>(score)
    s = re.sub(r"<<\s*(.*?)\s*>>\(\d+(?:\.\d+)?\)", r"\1", s, flags=re.DOTALL)
    # Quita <<...>>
    s = drop_angle_markers(s)
    # Decodifica placeholders si procede
    if decode_placeholders:
        s = decode_whitespace_placeholders(s)
    return s.strip()

def robust_parse_llm_json_pos_only(raw: str) -> Dict[str, Any]:
    """
    Parser robusto para {"explanation","activating_sentences"}.
    """
    s = strip_code_fences(normalize_quotes(raw))
    s_focus = s[s.find("{"): s.rfind("}") + 1] if "{" in s and "}" in s else s
    s_focus = remove_trailing_commas(s_focus)
    s_focus = re.sub(r",\s*\]", "]", s_focus)
    s_focus = re.sub(r",\s*\}", "}", s_focus)

    try:
        d = json.loads(s_focus)
        d = coerce_json_dict_pos_only(d)
    except Exception:
        expl = extract_string_field(s_focus, "explanation") or ""
        pos = extract_array_of_strings(s_focus, "activating_sentences") \
              or extract_array_of_strings(s_focus, "positives") or []
        d = coerce_json_dict_pos_only({"explanation": expl, "activating_sentences": pos})

    d["activating_sentences"] = clamp_k(d.get("activating_sentences", []), 5)
    d["explanation"] = (d.get("explanation") or "").strip()
    return d


# =============================================================================
# Lectura de resultados y flags de whitespace
# =============================================================================

def load_results_jsonl(path: str | Path) -> List[Dict[str, Any]]:
    items: List[Dict[str, Any]] = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            if not line.strip():
                continue
            items.append(json.loads(line))
    return items

def examples_need_whitespace(examples: List[str]) -> bool:
    """
    Usa los 'examples' (con <<...>>) para inferir si el feature depende de whitespace.
    Recibe lista de strings.
    """
    pat = re.compile(r"<<(.*?)>>", re.DOTALL)
    for text in examples:
        for inner in pat.findall(text or ""):
            if any(tok in inner for tok in ("\n", "\\n", "\t", "\\t", "[NL]", "[TAB]", "[WSP]")):
                return True
    return False


# =============================================================================
# Tokenización por lotes y activaciones
# =============================================================================

def batchify(xs: List[Any], batch_size: int) -> Iterable[List[Any]]:
    for i in range(0, len(xs), batch_size):
        yield xs[i:i + batch_size]

def tokenize_batch(sentences: List[str], tokenizer, pad_token_id: int) -> tuple[mx.array, List[int]]:
    ids: List[List[int]] = [[0] + tokenizer.encode(s) for s in sentences]
    lengths: List[int] = [len(x) for x in ids]
    T = max(lengths) if lengths else 1
    pad_id = pad_token_id
    padded = []
    for x in ids:
        if len(x) < T:
            padded.append(x + [pad_id] * (T - len(x)))
        else:
            padded.append(x)
    input_ids = mx.array(padded)  # [B, T]
    return input_ids, lengths

def get_hidden_states_batch(input_ids: mx.array, layer: int) -> mx.array:
    """
    Residual stream hasta 'layer' (incluida) para un batch [B,T].
    """
    h = model.model.embed_in(input_ids)
    for i, block in enumerate(model.model.h):
        h = block(h, mask="causal")
        if i == layer:
            break
    return h  # [B, T, D]

def max_feature_activation_per_sentence(
    sae: SparseAutoencoder,
    residual: mx.array,           # [B, T, D]
    feature_idx: int,
    lengths: List[int],
) -> np.ndarray:
    """
    Máximo del feature por secuencia sobre tokens reales (ignora padding).
    """
    x = sae.clip_norms(residual)     # [B,T,D]
    z = sae.encoder(x)               # [B,T,F]
    f = z[:, :, feature_idx]         # [B,T]
    B, T = f.shape
    out = np.zeros(B, dtype=np.float32)
    for b in range(B):
        L = max(lengths[b], 1)
        # ignorar posición 0 si corresponde a BOS
        out[b] = float(mx.max(f[b, 1:L]))
    return out

def evaluate_sentences_batch(
    sentences: List[str],
    sae: SparseAutoencoder,
    feature_idx: int,
    layer: int = TARGET_LAYER,
    batch_size: int = BATCH_SIZE,
) -> np.ndarray:
    vals: List[float] = []
    pad_id = tokenizer.pad_token_id
    for chunk in batchify(sentences, batch_size):
        input_ids, lengths = tokenize_batch(chunk, tokenizer, pad_id)
        residual = get_hidden_states_batch(input_ids, layer)
        max_vals = max_feature_activation_per_sentence(sae, residual, feature_idx, lengths)
        vals.extend(max_vals.tolist())
    return np.array(vals, dtype=np.float32)


# =============================================================================
# Métricas por feature (solo positivos)
# =============================================================================

def pos_metrics_at_tau(pos_vals: np.ndarray, tau: float) -> Dict[str, float]:
    """
    Métricas básicas con solo positivos y τ:
    - pos_hit_rate: fracción de oraciones con activación ≥ τ
    - mean_act / median_act / p75_act / max_act
    """
    pos_vals = np.asarray(pos_vals, dtype=np.float32)
    if pos_vals.size == 0:
        return {
            "n_pos": 0,
            "pos_hit_rate": np.nan,
            "mean_act": np.nan,
            "median_act": np.nan,
            "p75_act": np.nan,
            "max_act": np.nan,
        }
    hits = (pos_vals >= float(tau)).astype(np.int32)
    return {
        "n_pos": int(pos_vals.size),
        "pos_hit_rate": float(hits.mean()),
        "mean_act": float(np.mean(pos_vals)),
        "median_act": float(np.median(pos_vals)),
        "p75_act": float(np.percentile(pos_vals, 75.0)),
        "max_act": float(np.max(pos_vals)),
    }


# =============================================================================
# Carga y agrupación (detecta topk presentes)
# =============================================================================

def load_and_group_results(path_jsonl: str | Path) -> Dict[str, Dict[int, List[Dict[str, Any]]]]:
    """
    Devuelve estructura: grouped[sae_label][k] -> lista de entradas.
    Cada entrada: {
      "feature": int, "topk": int,
      "explanation": str, "examples": List[str], "activating_sentences": List[str],
      "ws_flag": bool
    }
    """
    raw_items = load_results_jsonl(path_jsonl)

    grouped: Dict[str, Dict[int, List[Dict[str, Any]]]] = {lab: defaultdict(list) for lab in SAE_LABELS}

    for item in raw_items:
        sae_model = str(item.get("sae_model", ""))
        if sae_model not in grouped:
            # Ignora SAEs no esperados (o podría añadirse dinámicamente)
            continue

        k = int(item.get("topk", -1))
        feat = int(item.get("feature_id"))

        # Parseo robusto del JSON del LLM (solo positivos)
        parsed = robust_parse_llm_json_pos_only(item.get("raw", ""))

        # ¿El feature depende de whitespace? (deducido de examples guardados)
        examples_list: List[str] = list(item.get("examples", []))
        ws_flag = examples_need_whitespace(examples_list)

        explanation = limpiar_salida_llm(parsed.get("explanation", ""), decode_placeholders=False)
        activating = [limpiar_salida_llm(s, decode_placeholders=ws_flag) for s in parsed.get("activating_sentences", [])]

        grouped[sae_model][k].append({
            "feature": feat,
            "topk": k,
            "explanation": explanation,
            "examples": examples_list,
            "activating_sentences": activating,
            "ws_flag": ws_flag
        })

    return {sae: dict(kmap) for sae, kmap in grouped.items()}


# =============================================================================
# τ_k por (SAE, k) desde scan_state
# =============================================================================

def load_tau_map_for_sae(sae_label: str) -> Dict[int, float]:
    """
    Carga scan_state_{sae_label}.pkl y devuelve {k: tau_k}.
    """
    scan_path = DATA_DIR / f"scan_state_{sae_label}.pkl"
    with open(scan_path, "rb") as f:
        state = json.load(f) if scan_path.suffix == ".json" else __import__("pickle").load(f)
    taus = state.get("taus", {})
    return {int(k): float(v) for k, v in taus.items()}


# =============================================================================
# Scoring por grupo (usa τ_k por grupo)
# =============================================================================

def score_group_entries(
    entries: List[Dict[str, Any]],
    sae: SparseAutoencoder,
    tau_k: float,
    batch_size: int = BATCH_SIZE,
) -> List[Dict[str, Any]]:
    """
    Métricas por feature (mismo SAE, mismo k), usando τ_k del grupo.
    """
    results: List[Dict[str, Any]] = []

    for item in tqdm(entries, desc="Scoring features"):
        f_idx = item["feature"]
        pos_texts: List[str] = item.get("activating_sentences", [])
        ws_flag: bool = item.get("ws_flag", False)

        if not pos_texts:
            results.append({
                "feature": f_idx, "topk": item["topk"], "tau": float(tau_k),
                "n_pos": 0,
                "pos_hit_rate": np.nan,
                "mean_act": np.nan, "median_act": np.nan, "p75_act": np.nan, "max_act": np.nan,
            })
            continue

        # Activaciones sobre las frases generadas por el LLM
        pos_vals = evaluate_sentences_batch(pos_texts, sae, feature_idx=f_idx,
                                            layer=TARGET_LAYER, batch_size=batch_size)

        # Métricas a τ_k
        m = pos_metrics_at_tau(pos_vals, tau_k)
        results.append({
            "feature": f_idx, "topk": item["topk"], "tau": float(tau_k), **m
        })

    return results


# =============================================================================
# Resumen, comparaciones y export
# =============================================================================

def summarize_scores(scores: List[Dict[str, Any]]) -> Dict[str, float]:
    def arr(key): return np.array([x[key] for x in scores], dtype=float)
    out = {
        "mean_pos_hit": float(np.nanmean(arr("pos_hit_rate"))) if scores else np.nan,
        "mean_mean_act": float(np.nanmean(arr("mean_act"))) if scores else np.nan,
        "mean_median_act": float(np.nanmean(arr("median_act"))) if scores else np.nan,
        "mean_p75_act": float(np.nanmean(arr("p75_act"))) if scores else np.nan,
        "mean_max_act": float(np.nanmean(arr("max_act"))) if scores else np.nan,
    }
    return out


# ---- Welch t-test con corrección por población finita ----

class TTestFPCResult(NamedTuple):
    statistic: float
    pvalue: float
    df: float
    n1: int
    n2: int

def welch_ttest_ind_fpc(
    a: np.ndarray,
    b: np.ndarray,
    N_pop: int | None = None,
) -> TTestFPCResult:
    """
    Welch t-test para dos muestras independientes con corrección por
    población finita (FPC) en cada grupo.

    - a, b: vectores de métricas (pos_hit_rate, mean_act, etc.)
    - N_pop: tamaño de la población finita (por ejemplo 2048 features).
             Si es None, no aplica FPC (equivalente a Welch estándar).
    """
    x = np.asarray(a, dtype=float)
    y = np.asarray(b, dtype=float)

    # Eliminamos NaN (equivalente a nan_policy="omit")
    x = x[np.isfinite(x)]
    y = y[np.isfinite(y)]

    n1, n2 = x.size, y.size
    if n1 < 2 or n2 < 2:
        return TTestFPCResult(np.nan, np.nan, np.nan, n1, n2)

    m1 = float(x.mean())
    m2 = float(y.mean())
    s1_sq = float(x.var(ddof=1))
    s2_sq = float(y.var(ddof=1))

    # Factor de corrección por población finita:
    # Var(ȳ) = (σ² / n) * (N - n) / (N - 1)
    if N_pop is not None and N_pop > 1:
        f1_sq = (N_pop - n1) / (N_pop - 1)
        f2_sq = (N_pop - n2) / (N_pop - 1)
        f1_sq = max(f1_sq, 0.0)
        f2_sq = max(f2_sq, 0.0)
    else:
        f1_sq = 1.0
        f2_sq = 1.0

    v1 = s1_sq / n1 * f1_sq
    v2 = s2_sq / n2 * f2_sq

    se = np.sqrt(v1 + v2)
    if se == 0.0:
        return TTestFPCResult(np.nan, np.nan, np.nan, n1, n2)

    t_stat = (m1 - m2) / se

    # Grados de libertad de Welch con varianzas ajustadas
    df_num = (v1 + v2) ** 2
    df_den = (v1 ** 2) / (n1 - 1) + (v2 ** 2) / (n2 - 1)
    df = df_num / df_den if df_den > 0 else np.nan

    if np.isfinite(df):
        pval = float(2 * stats.t.sf(np.abs(t_stat), df))
    else:
        pval = np.nan

    return TTestFPCResult(float(t_stat), pval, float(df), n1, n2)


def compare_groups(
    a: List[Dict[str, Any]],
    b: List[Dict[str, Any]],
    N_pop: int = POP_SIZE_FEATURES,
) -> Dict[str, Any]:
    """
    Welch t-test sobre métricas clave (solo-positivos),
    con corrección por población finita.

    - N_pop: tamaño de la población finita por grupo (features totales).
    """
    def arr(scores, key):
        return np.array([x[key] for x in scores], dtype=float)

    tests: Dict[str, Any] = {}

    # Métricas continuas
    for key in ["pos_hit_rate", "mean_act", "median_act", "p75_act", "max_act"]:
        ta = arr(a, key)
        tb = arr(b, key)
        tests[f"t_{key}"] = welch_ttest_ind_fpc(ta, tb, N_pop=N_pop)

    # Indicador binario "alguna oración ≥ τ" (anyhit)
    any_a = np.array(
        [float(x["pos_hit_rate"] > 0) if np.isfinite(x["pos_hit_rate"]) else np.nan for x in a],
        dtype=float,
    )
    any_b = np.array(
        [float(x["pos_hit_rate"] > 0) if np.isfinite(x["pos_hit_rate"]) else np.nan for x in b],
        dtype=float,
    )
    tests["t_anyhit"] = welch_ttest_ind_fpc(any_a, any_b, N_pop=N_pop)

    return tests


def export_feature_scores_csv(path: str | Path, group_name: str, scores: List[Dict[str, Any]]):
    fieldnames = [
        "group", "feature", "topk", "tau", "n_pos",
        "pos_hit_rate", "mean_act", "median_act", "p75_act", "max_act",
    ]
    with open(path, "w", encoding="utf-8", newline="") as f:
        w = csv.DictWriter(f, fieldnames=fieldnames)
        w.writeheader()
        for x in scores:
            row = {
                "group": group_name,
                "feature": x["feature"],
                "topk": x["topk"],
                "tau": x["tau"],
                "n_pos": x["n_pos"],
                "pos_hit_rate": x["pos_hit_rate"],
                "mean_act": x["mean_act"],
                "median_act": x["median_act"],
                "p75_act": x["p75_act"],
                "max_act": x["max_act"],
            }
            w.writerow(row)


# =============================================================================
# Ejecución
# =============================================================================

if __name__ == "__main__":
    if not RESULTS_JSONL.exists():
        raise FileNotFoundError(f"No existe {RESULTS_JSONL}")

    # 1) Cargar, parsear y agrupar por (SAE, k) presentes en llm-responses.jsonl
    grouped = load_and_group_results(RESULTS_JSONL)

    # 2) τ_k por (SAE, k) desde scan_state
    tau_map_by_sae: Dict[str, Dict[int, float]] = {}
    for sae_label in SAE_LABELS:
        tau_map_by_sae[sae_label] = load_tau_map_for_sae(sae_label)

    # 3) Scoring por grupo (SAE, k)
    scores_by_group: Dict[Tuple[str, int], List[Dict[str, Any]]] = {}
    ks_present_global: DefaultDict[int, int] = defaultdict(int)

    for sae_idx, sae_label in enumerate(SAE_LABELS):
        kmap = grouped.get(sae_label, {})
        for k, entries in kmap.items():
            if not entries:
                continue
            tau_k = tau_map_by_sae.get(sae_label, {}).get(int(k), None)
            if tau_k is None:
                print(f"[WARN] No hay τ_k para {sae_label} k={k}; se omite este grupo.")
                continue

            print(f"\n[INFO] Scoring {sae_label} @ k={k}  (features: {len(entries)})  τ={tau_k:.6f}")
            scores = score_group_entries(entries, SAES[sae_idx], tau_k=float(tau_k), batch_size=BATCH_SIZE)
            scores_by_group[(sae_label, k)] = scores
            ks_present_global[k] += 1

            # Resumen del grupo
            summary = summarize_scores(scores)
            print(f"\n== {sae_label} (k={k}) summary ==")
            for key, val in summary.items():
                print(f"{key:>15}: {val:.4f}" if np.isfinite(val) else f"{key:>15}: NaN")

            # Export opcional a CSV por grupo
            export_feature_scores_csv(DATA_DIR / f"{sae_label}__k{k}_feature_scores.csv",
                                      f"{sae_label}__k{k}", scores)

    # 4) Resumen agregando todos los k por SAE
    for sae_label in SAE_LABELS:
        all_scores = []
        for (lab, k), vals in scores_by_group.items():
            if lab == sae_label:
                all_scores.extend(vals)
        if all_scores:
            summary_all = summarize_scores(all_scores)
            print(f"\n== {sae_label} (ALL k) summary ==")
            for key, val in summary_all.items():
                print(f"{key:>15}: {val:.4f}" if np.isfinite(val) else f"{key:>15}: NaN")
            export_feature_scores_csv(DATA_DIR / f"{sae_label}__ALLk_feature_scores.csv",
                                      f"{sae_label}__ALLk", all_scores)

    # 5) Comparaciones: TODOS los SAEs contra el primero (baseline) por cada k
    if len(SAE_LABELS) >= 2:
        BASE_LABEL = SAE_LABELS[0]
        print(f"\n[INFO] Usando {BASE_LABEL} como baseline para comparaciones t-test.")

        for k in sorted(ks_present_global.keys()):
            base_scores = scores_by_group.get((BASE_LABEL, k), [])
            if not base_scores:
                continue

            for other_label in SAE_LABELS[1:]:
                other_scores = scores_by_group.get((other_label, k), [])
                if not other_scores:
                    continue

                print(f"\n== Welch t-tests ({other_label} vs {BASE_LABEL}) @ k={k} ==")
                tests = compare_groups(base_scores, other_scores)
                for name, res in tests.items():
                    print(f"{name:>15}: statistic={res.statistic:.4f}, pvalue={res.pvalue:.4g}")

        # 6) Comparación agregada ALL k: cada SAE vs baseline
        base_all = []
        for (lab, k), vals in scores_by_group.items():
            if lab == BASE_LABEL:
                base_all.extend(vals)

        for other_label in SAE_LABELS[1:]:
            other_all = []
            for (lab, k), vals in scores_by_group.items():
                if lab == other_label:
                    other_all.extend(vals)

            if base_all and other_all:
                print(f"\n== Welch t-tests ({other_label} vs {BASE_LABEL}) @ ALL k ==")
                tests = compare_groups(base_all, other_all)
                for name, res in tests.items():
                    print(f"{name:>15}: statistic={res.statistic:.4f}, pvalue={res.pvalue:.4g}")


# =============================================================================
# Gráfico único: Δ mean_pos_hit vs baseline (primer SAE), con ejes invertidos
# =============================================================================
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import re

def parse_ranks(stem):
    m = re.search(r"Er(\d+)_Dr(\d+)", stem)
    if not m:
        return (0,0)
    return int(m.group(1)), int(m.group(2))

def pretty_label(stem):
    rE, rD = parse_ranks(stem)
    return f"LRQ-SAE ($r_{{enc}}={rE}$, $r_{{dec}}={rD}$)"

metric = "pos_hit_rate"

# 1) Recolectar las métricas por-feature para ALL k
metric_values = {}
for stem in SAE_LABELS:
    vals = []
    for (lab, k), groupvals in scores_by_group.items():
        if lab == stem:
            for x in groupvals:
                v = x.get(metric, np.nan)
                if np.isfinite(v):
                    vals.append(v)
    vals = np.array(vals, dtype=float)
    vals = vals[np.isfinite(vals)]
    if vals.size > 0:
        metric_values[stem] = vals

# Debe haber baseline y otros modelos
valid_stems = [s for s in SAE_LABELS if s in metric_values]
if len(valid_stems) < 2:
    print("[WARN] No hay suficientes modelos con datos.")
else:
    BASE = valid_stems[0]
    base_vals = metric_values[BASE]
    n0 = base_vals.size
    mean0 = base_vals.mean()
    std0 = base_vals.std(ddof=1)
    se0 = std0 / np.sqrt(n0)

    other_stems = valid_stems[1:]

    diff_means = []
    diff_err = []
    labels = []

    for stem in other_stems:
        arr = metric_values[stem]
        n1 = arr.size
        mean1 = arr.mean()
        std1 = arr.std(ddof=1)
        se1 = std1 / np.sqrt(n1)

        diff = mean1 - mean0
        diff_se = np.sqrt(se0**2 + se1**2)
        df = max(min(n0-1, n1-1), 1)
        tcrit = stats.t.ppf(0.975, df)
        halfwidth = tcrit * diff_se

        diff_means.append(diff)
        diff_err.append(halfwidth)
        labels.append(pretty_label(stem))

    diff_means = np.array(diff_means)
    diff_err = np.array(diff_err)
    y = np.arange(len(other_stems))

    # 2) Gráfico con ejes invertidos
    plt.figure(figsize=(8, 3))
    ax = plt.gca()

    # Línea punteada vertical con label
    ax.axvline(0.0, color="gray", linestyle="--", linewidth=1.2, label="Baseline (Δ = 0)")

    # Puntos con IC
    ax.errorbar(diff_means, y, xerr=diff_err, fmt="o", capsize=4, color="black")

    ax.set_yticks(y)
    ax.set_yticklabels(labels)

    ax.set_xlabel(r"$\Delta\,pos_{hit}$")
    ax.legend()

    plt.tight_layout()
    plt.show()
````
